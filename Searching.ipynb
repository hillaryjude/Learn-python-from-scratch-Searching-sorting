# Linear search

```python
def search(arr,key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1
        
```

```python
arr = [1,2,3,4,5,6,7,9]
print(search(arr,6))
```

```python
print(search(arr,8))
```

# Binary Search

The below code works only when the array is sorted in ascending order.

```python
def binary(arr,key,n):
    l = 0
    r = n-1
    while l<=r:
        m = (l+r)//2
        if key == arr[m]:
            return m
        elif key < arr[m]:
            r = m-1
        else:
            l = m+1
    return -1
```

```python
arr = [1,2,3,4,5,6,7]
n = len(arr)
print(binary(arr,5,n))
```

# Recursive Binary search

```python
def recbinary(arr,l,r,key):
    if l>r:
        return -1
    mid = (l+r)//2
    if key == arr[mid]:
        return mid
    elif key > arr[mid]:
        return recbinary(arr,mid+1,r,key)
    else:
        return recbinary(arr,l,mid-1,key)   
```

```python
arr = [1,2,3,4,5,6,7]
n= len(arr)
print(recbinary(arr,0,n-1,4))
```

# Sqrt of a number

```python
def sqrt(num):
    if num == 0 or num ==1:
        return num
    l = 2
    r = num//2
    while l<=r:
        mid = (l+r)//2
        temp = mid * mid
        if temp == num:
            return mid
        elif temp > num:
            l = mid+1
        else:
            r = mid-1
    return r
```

```python
print(sqrt(4))
```

```python
sqrt(100)
```
